```python
import pandas as pd

all_stock_data = pd.read_parquet('/kaggle/input/9000-tickers-of-stock-market-data-full-history/all_stock_data.parquet')
```

```python
all_stock_data
```

```python
len(all_stock_data['Ticker'].unique())
```

```python
all_stock_data.groupby(by='Ticker').size().sort_values()
```

```python
all_stock_data[all_stock_data['Ticker'].values=='AAPL'].set_index('Date')['Close'].plot(title='AAPL Close Price (Daily)')
```

```python
stock_ticker = 'AAPL'
window_size_small = 50
window_size_large = 200
days_of_history = 252

single_stock_price = all_stock_data[all_stock_data['Ticker'].values==stock_ticker].set_index('Date')[['Close']]

single_stock_price[f'Rolling Mean {window_size_small}'] = single_stock_price['Close'].rolling(window=window_size_small).mean()

single_stock_price[f'Rolling Mean {window_size_large}'] = single_stock_price['Close'].rolling(window=window_size_large).mean()

single_stock_price.tail(days_of_history).plot(title=f'1Y {stock_ticker} Stock Close Price (Daily)')
```
